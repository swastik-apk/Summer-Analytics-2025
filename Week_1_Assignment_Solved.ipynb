### 1. What is name of car that has the highest horsepower?  

```python
#your code here
new_data = pd.read_csv("Cars.csv")

# Find the row with the maximum horsepower
car_with_highest_horsepower = new_data.loc[new_data['horsepower'].idxmax()]

# Get the name of the car
car_name = car_with_highest_horsepower['name']

# Print the name of the car
print(car_name)
```

### 2. How many cars have mpg ≥ 35?

```python
#your code here

# Filter the DataFrame to include cars with mpg >= 35
cars_mpg_ge_35 = new_data[new_data['mpg'] >= 35]

# Get the count of such cars
num_cars = len(cars_mpg_ge_35)

# Print the result
print(f"Number of cars with mpg \u2265 35: {num_cars}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here

# Filter the DataFrame based on horsepower > 100 and weight < 3000
filtered_cars = new_data[(new_data['horsepower'] > 100) & (new_data['weight'] < 3000)]

# Get the value counts for the 'origin' column in the filtered DataFrame
origin_counts = filtered_cars['origin'].value_counts()

# Get the most common origin
most_common_origin = origin_counts.index[0]

# Print the most common origin
print(f"The most common origin for cars with horsepower > 100 and weight < 3000 is: {most_common_origin}")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here

# Filter the DataFrame to include cars from 'japan'
japan_cars = new_data[new_data['origin'] == 'japan']

# Calculate the mean acceleration for cars from Japan
mean_acceleration_japan = japan_cars['acceleration'].mean()

# Round the mean acceleration to 2 decimal places
rounded_mean_acceleration = round(mean_acceleration_japan, 2)

# Print the result
print(f"The mean acceleration of cars from Japan is: {rounded_mean_acceleration}")
```

### 5. Which year had the highest average mpg?

```python
#your code here

# Group by 'model_year' and calculate the mean of 'mpg' for each year
average_mpg_by_year = new_data.groupby('model_year')['mpg'].mean()

# Find the year with the highest average mpg
year_highest_avg_mpg = average_mpg_by_year.idxmax()

# Print the result
print(f"The year with the highest average mpg is: {year_highest_avg_mpg}")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here

# Calculate the median mpg
median_mpg = new_data['mpg'].median()

# Filter for cars with mpg above the median
above_median_mpg_cars = new_data[new_data['mpg'] > median_mpg].copy()

# Calculate the horsepower to weight ratio for these cars
above_median_mpg_cars['hp_to_weight_ratio'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']

# Find the maximum ratio
max_ratio = above_median_mpg_cars['hp_to_weight_ratio'].max()

# Find the car(s) with the maximum ratio
best_ratio_cars = above_median_mpg_cars[above_median_mpg_cars['hp_to_weight_ratio'] == max_ratio]

# Print the name(s) of the car(s)
print("Car(s) with the best horsepower to weight ratio among those with above-median mpg:")
for car_name in best_ratio_cars['name']:
    print(car_name)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here

# Group by 'model_year' and 'origin' and calculate the mean of 'mpg'
average_mpg_evolution = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Set up the plot style
sns.set_style("whitegrid")

# Create the multi-line plot
plt.figure(figsize=(12, 7))

# Get unique origins
origins = average_mpg_evolution['origin'].unique()

# Plot a line for each origin
for origin in origins:
    subset = average_mpg_evolution[average_mpg_evolution['origin'] == origin]
    plt.plot(subset['model_year'], subset['mpg'], marker='o', label=origin)

# Add labels and title
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.title("Evolution of Average MPG by Origin Over Years")
plt.legend(title="Origin")
plt.grid(True)
plt.xticks(average_mpg_evolution['model_year'].unique()) # Ensure all model years are shown as ticks
plt.tight_layout()

# Save the plot to a file
plt.savefig("average_mpg_evolution_by_origin.png")

# Display the plot
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here

# Load the CSV file into a DataFrame
df = pd.read_csv("Cars.csv")

# 1. Identify car names that appear in more than one model_year
# Group by 'name' and count unique 'model_year' values
car_years_count = df.groupby('name')['model_year'].nunique()

# Filter for cars appearing in more than one year
multi_year_cars = car_years_count[car_years_count > 1].index

# Filter the original DataFrame to include only multi-year cars
filtered_df = df[df['name'].isin(multi_year_cars)]

# 2. For each such name, compute the standard deviation of mpg across years
#    Also compute number of appearances and average mpg
consistent_cars_data = filtered_df.groupby('name').agg(
    std_mpg=('mpg', 'std'),
    num_appearances=('name', 'count'),
    avg_mpg=('mpg', 'mean')
).reset_index()

# 3. Filter for consistent cars: at least 2 appearances and std(mpg) < 1.0
consistent_cars = consistent_cars_data[
    (consistent_cars_data['num_appearances'] >= 2) &
    (consistent_cars_data['std_mpg'] < 1.0)
]

# If there are no consistent cars, handle this case
if consistent_cars.empty:
    print("No consistent car models found based on the given criteria.")
else:
    # 4. Find the car(s) with the lowest variation in mpg among consistent cars
    #    This is implicitly handled by filtering for std_mpg < 1.0
    #    The request asks for the car(s) with *the lowest variation* among those,
    #    so we'll find the minimum std_mpg within the already filtered 'consistent_cars'
    min_std_mpg_among_consistent = consistent_cars['std_mpg'].min()
    lowest_variation_cars = consistent_cars[consistent_cars['std_mpg'] == min_std_mpg_among_consistent]

    # 5. Report the model name(s), number of appearances, and the average mpg
    #    Bonus: Sort the result by number of appearances (descending), then mpg (descending).
    final_report = lowest_variation_cars[['name', 'num_appearances', 'avg_mpg', 'std_mpg']].sort_values(
        by=['num_appearances', 'avg_mpg'],
        ascending=[False, False]
    )

    print("Consistent car model(s) with the lowest variation in mpg:")
    print(final_report.to_string(index=False))
```
